```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Data
X = np.array([2, 3, 7]).reshape(-1, 1)
y = np.array([-1, -1, 1])

# SVM with linear kernel
model = SVC(kernel='linear')
model.fit(X, y)

# Support vectors
support_vectors = model.support_vectors_



# Computing the decision boundary
# For a linear SVM, the decision boundary in 1D is a point x such that w*x + b = 0
w = model.coef_[0]
b = model.intercept_
decision_boundary = -b / w

# Plotting with the decision boundary
plt.figure(figsize=(8, 4))
plt.scatter(X, np.zeros_like(X), c=y, marker='o', label='Data Points')
plt.scatter(support_vectors, np.zeros_like(support_vectors), facecolors='none', edgecolors='k', label='Support Vectors')
plt.axvline(x=decision_boundary, color='red', linestyle='-', label='Decision Boundary')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.title('1D SVM with Linear Kernel and Decision Boundary')
plt.legend()
plt.show()

```
