```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg as linalg
```

```python
x = np.array([2, 3, 4])
y = np.array([4, 6, 8])

N = x.shape[0]
print(x)
print(y)
print(N)
```

```python
data_range = np.linspace(start=0, stop=20, num=201)
```

```python
plt.figure(figsize = (6, 3))
plt.plot(x, y, "k.", markersize = 10)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
# calculate A
A = np.array([[N, np.sum(x)],
              [np.sum(x), np.sum(x**2)]])
print(A)
```

```python
# calculate b
b = np.array([[np.sum(y)], [np.sum(y * x)]])
print(b)
```

```python
# calculate w
w = np.matmul(linalg.cho_solve(linalg.cho_factor(A), np.eye(2)), b)
print(w)
```

```python
y_hat_test = np.matmul(np.stack((np.repeat(1.0, len(x)), x)).T, w)

plt.figure(figsize = (10, 6))
plt.plot(x, y, "k.", markersize = 10)
plt.plot(x, y_hat_test, "b-")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
